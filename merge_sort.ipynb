```python
import numpy as np
```

```python
def merge_sort(lst):
    num_elems = len(lst)
    
    if num_elems <= 1:
        return lst
    
    half = num_elems // 2
    
    first, second = merge_sort(lst[:half]), merge_sort(lst[half:])
    i, j = 0, 0
    merged = []
    for k in range(num_elems):
        if i >= half:
            merged.extend(second[j:])
            break
        elif j >= num_elems - half:
            merged.extend(first[i:])
            break
        if first[i] < second[j]:
            merged.append(first[i])
            i += 1
        else:
            merged.append(second[j])
            j += 1
            
    return merged
```

```python
max_num = 10
for i in range(16):
    arr = np.random.randint(max_num, size=30).tolist()
    sorted_arr = merge_sort(arr)
    print(arr)
    print(sorted_arr)
    print("")
    assert sorted_arr == sorted(arr)
```
