```python
import random
import numpy as np
```

```python
def quick_sort(lst, start, end):
    """
    in-place implementation of quick sort
    """
    num_elems = end - start
    if num_elems > 1: # (otherwise do nothing) 
        pivot_idx = random.randint(start, end - 1)
        lst[start], lst[pivot_idx] = lst[pivot_idx], lst[start] # swap pivot element to position 0
        pivot = lst[start]
        i = start + 1
        for j in range(i, end):
            if lst[j] < pivot:
                lst[j], lst[i] = lst[i], lst[j] # swap
                i += 1
        lst[start], lst[i - 1] = lst[i - 1], lst[start] # position pivot at the "center"

        # in-place sorts
        quick_sort(lst, start, i - 1)
        quick_sort(lst, i, end)
```

```python
max_num = 10
for i in range(16):
    arr = np.random.randint(max_num, size=30).tolist()
    print(arr)
    quick_sort(arr, 0, len(arr))
    print(arr)
    print("")
    assert arr == sorted(arr)
```
